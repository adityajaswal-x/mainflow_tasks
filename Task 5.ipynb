```python
#Let's begin formulating queries for this dataset.

#Q.1 How many people have heart disease and how many people doesn't have heart disease ?

#getting the values
df.target.value_counts()

#plotting bar chart
df.target.value_counts().plot(kind = 'bar', color =["orchid", "salmon"])
plt.title("Heart Disease values")
plt.xlabel("1 = Heart Disease, 0 = No heart Disease")
plt.ylabel("Amount");
```

```python
#plotting a pie chart
df.target.value_counts().plot(kind = 'pie', figsize = (8, 6))
plt.legend(["Disease", "No disease"]);
```

```python
# '0' represents 'Female'
# '1' represents 'Male'
# SEX column part

# '0' represents 'No disease'
# '1' represents 'Disease'
# Target column part

# Now let's check how many 'Male' and 'Female' are in the dataset
df.sex.value_counts()

# Plotting a pie chart
df.sex.value_counts().plot(kind='pie', figsize=(8, 6))
plt.title('Male Female ratio')
plt.legend(['Male', 'Female']);
```

```python
#Q2. People of which sex has the most heart disease?
pd.crosstab(df.target, df.sex)
```

```python
sns.countplot(x = 'target', data = df, hue ='sex')
plt.title('Heart Disease Frequency for Sex')
plt.xlabel('0 = No heart Disease, 1 = Heart Disease');
```

```python
# Number of male is more than double in our dataset than female.
# More than 45% male has heart disease and 75% female has heart disease.
```

```python
# Q3. People of which sex has which type of chest pain most?

# Counting values for different chest pain
df.cp.value_counts()
```

```python
# Plotting a bar chart
df.cp.value_counts().plot(kind='bar', color=['salmon', 'lightskyblue', 'springgreen', 'khaki'])
plt.title('Chest pain type count');
```

```python
pd.crosstab(df.sex, df.cp)
```

```python
pd.crosstab(df.sex, df.cp).plot(kind='bar', color=['coral', 'lightskyblue', 'plum', 'khaki'])
plt.title('Type of chest pain for sex')
plt.xlabel('0 = Female, 1 = Male');
```

```python
# Now question 4!

# Q4. People with which chest pain are most prone to have heart disease?
pd.crosstab(df.cp, df.target)

sns.countplot(x='cp', data=df, hue='target');
```

```python
# Most of the people who has 'type 0' chest pain has less chance of heart disease.
# And we see the opposite for other types.

# Now let's take a look at our age column.
# Create a distribution plot with normal distribution curve
sns.displot(x='age', data=df, bins=30, kde=True);
```

```python
# '58-59' year old people are most in the dataset.

# Let's plot another distribution plot for 'Maximum heart rate'
sns.displot(x='thalach', data=df, bins=30, kde=True, color='chocolate');
```

```python
# From this plot we get a clear overview about Maximum heart rate represented by 'thalach'
```

```python
#Additional Questions with Answers

#Q5. What is the relationship between age and cholesterol levels in the dataset?

# Creating a scatter plot to examine the relationship between age and cholesterol levels
sns.scatterplot(x='age', y='chol', data=df, hue='target', palette='viridis');
plt.title('Age vs Cholesterol Levels');
plt.xlabel('Age');
plt.ylabel('Cholesterol Level (chol)');
```

```python
#Q6.  How does the type of chest pain (cp) correlate with the maximum heart rate achieved (thalach)?

# Creating a boxplot to analyze the relationship between chest pain type and maximum heart rate
sns.boxplot(x='cp', y='thalach', data=df, palette='Set2');
plt.title('Chest Pain Type vs Maximum Heart Rate');
plt.xlabel('Chest Pain Type (cp)');
plt.ylabel('Maximum Heart Rate (thalach)');

```

```python
#Q7. How does the electrocardiographic result (restecg) relate to the presence of heart disease?

# Analyzing the relationship between 'restecg' and heart disease
pd.crosstab(df.restecg, df.target)

sns.countplot(x='restecg', data=df, hue='target');
plt.title('Heart Disease Frequency for Resting ECG Results');
plt.xlabel('Resting ECG Result (restecg)');
```
