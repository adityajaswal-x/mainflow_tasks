```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('householdtask3.csv')
```

```python
data.head(10)
```

```python
plt.scatter(data['year'], data['own'])

plt.title("Scatter Plot")

plt.xlabel('year')
plt.ylabel('own')

plt.show()
```

```python
plt.plot(data['year'])
plt.plot(data['own'])

plt.title("L")

plt.xlabel('year')
plt.ylabel('own')

plt.show() 
```

```python
plt.bar(data['year'], data['own'])

plt.title("Bar Chart")

plt.xlabel('year')
plt.ylabel('own')

plt.show() 
```

```python
plt.hist(data['income'])

plt.title("Histogram")

plt.show()
```
