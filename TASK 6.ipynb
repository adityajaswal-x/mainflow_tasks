```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Load the Diabetes Dataset
df = pd.read_csv("diabetes_data.csv")
```

```python
# Display the first few rows
print("Head of the dataset:")
df.head()
```

```python
# Display the last few rows
print("\nTail of the dataset:")
df.tail()
```

```python
# Summary statistics
print("\nDescriptive statistics:")
df.describe()
```

```python
# Check for null values
print("\nChecking for null values:")
df.isnull().sum()
```

```python
# Visualization 1: Distribution of Age
plt.figure(figsize=(8, 5))
sns.histplot(df['Age'], bins=20, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Visualization 2: BMI vs. Age scatter plot
plt.figure(figsize=(8, 5))
sns.scatterplot(x='Age', y='BMI', data=df)
plt.title('BMI vs. Age')
plt.xlabel('Age')
plt.ylabel('BMI')
plt.show()
```

```python
# Visualization 3: Box plot of BMI by Gender
plt.figure(figsize=(8, 5))
sns.boxplot(x='Gender', y='BMI', data=df)
plt.title('BMI by Gender')
plt.xlabel('Gender')
plt.ylabel('BMI')
plt.show()
```

```python
# Systolic BP Distribution
plt.figure(figsize=(8, 5))
sns.histplot(df['SystolicBP'], bins=20, kde=True)
plt.title('Systolic Blood Pressure Distribution')
plt.xlabel('Systolic BP')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='Ethnicity', data=df)
plt.title('Count of Ethnicity')
plt.xlabel('Ethnicity')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='Gender', hue='Smoking', data=df)
plt.title('Smoking Status by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='Gender', y='CholesterolTotal', data=df)
plt.title('Cholesterol by Gender')
plt.xlabel('Gender')
plt.ylabel('Total Cholesterol')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x='Age', y='PhysicalActivity', data=df)
plt.title('Physical Activity vs. Age')
plt.xlabel('Age')
plt.ylabel('Physical Activity')
plt.show()
```
